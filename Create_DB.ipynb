```python
import requests
import urllib
from bs4 import BeautifulSoup
from collections import defaultdict
import re

import olefile
import pprint
import pdftotext
```

```python
titles = []
with open("./movie_titles.txt", 'r', encoding='UTF8') as f:
    titles = f.readlines()
    titles = list(map(lambda x: x.strip(), titles))

pprint.pprint(titles)
    
data = defaultdict(list)
    
for movie in titles:
    
    # Split the file format
    file_format = movie.split('.')[1].strip()
    print("file format:", file_format)
    print(movie)
    
    
    # I. Text Format handling
    if (file_format == 'txt'):
        continue
        with open(movie, 'rt', encoding='UTF8') as f:
            while(True):
                line = f.readline()

                if (not line):
                    break

                if (':' in line):
                    character = line.split(':')[0].strip() # Why was it rstrip()??!
                    script = line.split(':')[1].strip()

#                     Re is not working properly -> need to look back later
#                     pattern = r'^\(.+\)'
#                     print(re.search(pattern, script))
#                     print(script)

                    while ('(' in script and ')' in script):
                        script = script.split('(')[0] + script.replace(script.split(')')[0] + ')', '')

                    key = character
                    data[key].append(script)
    
    # II. PDF Format Handling
    # Scanned pdf file might not work -> I think manual handling is enough
    elif (file_format == 'pdf'):
        with open(movie, 'rb') as f:
            fileReader = pdftotext.PDF(f)
            
            for page in fileReader:
                for line in page.split('\r\n'):
                    print(line, " -> and of the line")
#                     if (re.search(r'\s{11}', line) != None and re.search(r'\s{12}', line) == None):

#                         pattern = r'\s{11}'
#                         character = re.split(pattern, line)[0].strip()
#                         script = re.split(pattern, line)[1].strip()

#                         while ('(' in script and ')' in script):
#                             script = script.split('(')[0] + script.replace(script.split(')')[0] + ')', '')

#                         key = character
#                         data[key].append(script)

    # II. HWP Format Handling: NOT WORKING
    # HWP is not properly working with python environment
    # So it is rather better to use txt converted file as an input
    elif (file_format == 'hwp'):
        continue
        with olefile.OleFileIO(movie) as f:
            encoded_text = f.openstream('PreText').read()
            decoded_text = encoded_text.decode('UTF-8')
            
    else:
        print("Unknown format detected")
        print("-----")
        print("File Name:", file_format)
        print("-----")
    
pprint.pprint(data)
            
```
